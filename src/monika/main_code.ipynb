```python
df_sample = df.sample(n=5000, random_state=42)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)
```

```python
from langchain_core.documents import Document
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document

splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

documents = []

for idx, row in df_sample.iterrows():
    text = row["embedding_text"]
    if not isinstance(text, str) or not text.strip():
        continue

    chunks = splitter.split_text(text)

    for i, chunk in enumerate(chunks):
        documents.append(
            Document(
                page_content=chunk,
                metadata={
                    "row_id": idx,
                    "asin": row["asin"],
                    "parent_asin": row["parent_asin"],
                    "category": row["clean_categories"],
                    "chunk_id": i
                }
            )
        )
```

```python
len(documents)
```

```python
documents[0]
```

```python
vectorstore = FAISS.from_documents(
    documents,
    embeddings
)
```

```python
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})
```

```python
results = retriever.invoke("best waterfiler of good quality in less than 20000")
```

```python
def inspect_result(result, df):
    row_id = result.metadata["row_id"]
    row = df.loc[row_id]

    print("="*80)
    print("ASIN:", row["asin"])
    print("PRODUCT TITLE:", row["product_title"])
    print("CATEGORY:", row["categories"])
    print("PRICE:", row["price"])
    print("RATING:", row["rating"])
    print("-"*80)
    print("RETRIEVED CHUNK:")
    print(result.page_content[:800])
    print("-"*80)
    print("FULL CLEAN REVIEW TEXT:")
    print(row["clean_review_text"][:800])
    print("="*80)
```
